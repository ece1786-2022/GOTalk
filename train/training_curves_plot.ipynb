```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_loss_ls = [7.001300,5.125400,3.968200,3.614600,3.252600, 3.242000, 3.140900, 3.062200, 2.877700, 2.877600, 2.848400]
validation_loss_ls = [5.906636, 4.208858, 3.765974, 3.627266, 3.604956, 3.568900, 3.545990, 3.538033, 3.557756, 3.549932, 3.549331]


steps = []
for i in range(50, 551, 50):
    steps.append(i)
```

```python
# Traning plot
plt.figure(figsize=(10,6))
plt.plot(steps, train_loss_ls, 'g', label='Training loss')
plt.plot(steps, validation_loss_ls, 'b', label='Valid loss')
plt.title('Training Curves', size =15)
plt.xlabel('Steps', size = 12)
plt.ylabel('Loss', size = 12)
plt.legend()
plt.show()
```
