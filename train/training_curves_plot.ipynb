```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_loss_ls = [4.5479, 3.6704, 2.1481, 2.1394, 0.8494, 0.8318, 0.3462, 0.3471]
validation_loss_ls = [3.7060, 3.6118, 3.9275, 3.8731, 4.8428, 4.9033, 5.2662, 5.3089]
steps = []
for i in range(500, 4001, 500):
    steps.append(i)
```

```python
# Traning plot
plt.figure(figsize=(10,6))
plt.plot(steps, train_loss_ls, 'g', label='Training loss')
plt.plot(steps, validation_loss_ls, 'b', label='Valid loss')
plt.title('Training Curves', size =15)
plt.xlabel('Steps', size = 12)
plt.ylabel('Loss', size = 12)
plt.legend()
plt.show()
```
